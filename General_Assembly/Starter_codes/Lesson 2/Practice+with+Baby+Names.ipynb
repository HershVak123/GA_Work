```python
#Import dataframe
df = pd.read_csv("https://data.cityofnewyork.us/api/views/25th-nujf/rows.csv?accessType=DOWNLOAD")
```

```python
#How many babies were born in 2011-2014 with your name?
df_me = df[df["Child's First Name"] == "Hersh"]
df_me["Child's First Name"]
```

```python
#How many babies were born with one of the following names?
#Christina
#Alvin
#Keshav
#Alfredo
#Gino
#Brian
#Ryan
#Chuck
#Kiran
#Eva
#Amanjeet
#Meigang
#Graham

df_name = df[df["Child's First Name"] == "Graham"]
df_name["Count"].sum()

# We just change the value for the name in "Child's First Name" == X to find the count for every other name.
```

```python
#What were the top five most popular female names for babies with Hispanic mothers in 2011? 
df_ex1 = df[(df["Year of Birth"] == 2011) & (df["Ethnicity"] == "HISPANIC") & (df["Gender"] == "FEMALE")]
df_ex1.sort_values(by = "Count", ascending = False)
```

```python
#What were the top five most popular male names for babies with black non-Hispanic mothers in 2014?

df_ex2 = df[(df["Gender"] == "MALE") & (df["Ethnicity"] == "BLACK NON HISPANIC") & (df["Year of Birth"] == 2014)]
df_ex2.sort_values(by = ["Rank", "Child's First Name"])
```

```python
#What were the top five most popular names (including  males AND females in 2013?
df_ex3 = df[(df["Year of Birth"]==2013)]
df_ex3["Child's First Name"].value_counts().head(10)
```

```python
#What were the top five most popular three-letter baby names for girls with white non-Hispanic mothers in 2013?
df_ex4 = df[(df["Gender"] == "FEMALE") & (df["Ethnicity"] == "WHITE NON HISPANIC") & (df["Year of Birth"] == 2013)].drop_duplicates()
df_ex4["Less_Than_3"] = np.where((df_ex4["Child's First Name"].str.len() < 4), 1, 0)
df_ex4_2 = df_ex4[(df_ex4["Less_Than_3"] == 1)]
df_ex4_2.sort_values(by = "Count", ascending = False)
```
